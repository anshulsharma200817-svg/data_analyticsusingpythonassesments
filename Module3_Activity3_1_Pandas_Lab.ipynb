## Task 1: Data Creation & Loading (1 Mark)
Dataset:
    'Name': ['Aman','Riya','Rahul','Sneha','Karan'],
    'Department': ['IT','HR','IT','Finance','HR'],
    'Salary': [50000,60000,55000,65000,52000]

```python
# Create:
# 1. A Pandas Series from a list of 5 elements
# 2. A DataFrame with columns: Name, Department, Salary
# 3. Write a dataset using (any one): CSV / Excel / JSON
import pandas as pd
data = [10, 20, 30, 40, 50]
series = pd.Series(data)

print("Pandas Series:")
print(series)
data = {
    "Name": ["Aman", "Riya", "Rahul", "Sneha", "Karan"],
    "Department": ["IT", "HR", "IT", "FINANCE", "HR"],
    "Salary": [50000, 60000, 55000, 65000, 52000]
}

df = pd.DataFrame(data)

print("\nDataFrame:")
print(df)
df.to_csv("employees.csv", index=False)
df.to_excel("employees.xlsx", index=False)
df.to_json("employees.json", orient="records")

print("\nFiles saved successfully (CSV, Excel, JSON).")

```

## Task 2: Data Exploration (1 Mark)

```python
# Perform:
# Display first 3 rows
# Display dataset information
# Display statistical summary
print("First 3 Rows:")
print(df.head(3))
print("\nDataset Information:")
print(df.info())
print("\nStatistical Summary:")
print(df.describe())

```

## Task 3: Indexing, Filtering & Sorting (1 Mark)

```python
# Perform:
# 1. Extract first 3 rows using .loc[]
# 2. Extract Salary column using .iloc[]
# 3. Filter employees with Salary > 55000
# 4. Sort dataset by Salary (descending)

print("First 3 rows using loc:")
print(df.loc[0:2])

print("\nSalary column using iloc:")
print(df.iloc[:, 2])
print("\nEmployees with Salary > 55000:")
print(df[df["Salary"] > 55000])

# 4. Sort dataset by Salary (descending)
print("\nDataset sorted by Salary (descending):")
print(df.sort_values(by="Salary", ascending=False))
```

## Task 4: Group-by & Aggregation (1 Mark)

```python
# Perform:
# 1. Find average salary department-wise
# 2. Find total and count of employees per departmentimport pandas as pd
print("Average Salary Department-wise:")
avg_salary = df.groupby("Department")["Salary"].mean()
print(avg_salary)
print("\nTotal Salary and Employee Count Department-wise:")
dept_stats = df.groupby("Department")["Salary"].agg(['sum', 'count'])
print(dept_stats)
```

## Task 5: Data Cleaning & Time-Series (1 Mark)

```python
# Given:
# df['Joining_Date'] = ['2022-01-10','2021-07-15','2023-03-12','2020-11-05','2022-05-20']

# Perform:
# 1. Convert Joining_Date to datetime
# 2. Extract Year and Month
# 3. Identify duplicate rows
# 4. Remove duplicate rows
import pandas as pd
df['Joining_Date'] = ['2022-01-10','2021-07-15','2023-03-12','2020-11-05','2022-05-20']
df['Joining_Date'] = pd.to_datetime(df['Joining_Date'])

# 2. Extract Year and Month
df['Year'] = df['Joining_Date'].dt.year
df['Month'] = df['Joining_Date'].dt.month

print("DataFrame with Year and Month:")
print(df)
print("\nDuplicate Rows:")
duplicates = df[df.duplicated()]
print(duplicates)
df_cleaned = df.drop_duplicates()

print("\nDataFrame after removing duplicates:")
print(df_cleaned)
```
